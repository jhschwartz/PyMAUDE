```python
# TODO: make this into an example notebook! 

import sys
from pathlib import Path
import pandas as pd


# Add src to path for development
src_path = Path().resolve().parent.parent / 'src'
sys.path.insert(0, str(src_path))

from pymaude import MaudeDatabase, SelectionManager
from pymaude.selection_widget import SelectionWidget

db_path = 'maude_2008_2025.db'
widget_selection_path = 'search_selections.json'

db = MaudeDatabase(db_path, verbose = True)
db.info()

manager = SelectionManager('selection_widet', widget_selection_path, db_path)
widget = SelectionWidget(manager, db)
widget.display()
```

```python
results = widget.get_results()
```

```python
results.to_df()
```
